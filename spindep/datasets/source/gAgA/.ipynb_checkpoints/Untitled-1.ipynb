```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

# ── Paste the confirmed path here ──
data_dir =  "/mnt/c/Users/DELL/Documents/MSC RESEARCH/Data/Spin-Dependent-5th-Force-Limits-v3.2/Lei-Cong-Spin-Dependent-5th-Force-Limits-a604398/Dataset/normalized/gAgA/lepton-lepton"  # Update with your folder path


if not data_dir.endswith("/"):
    data_dir += "/"

files = glob.glob(data_dir + "*.csv")

if len(files) == 0:
    raise FileNotFoundError(f"No CSV files found in: {data_dir}")

datasets = {}
for f in files:
    name = os.path.basename(f).replace(".csv", "")
    try:
        df = pd.read_csv(f, header=None, names=["lambda_m", "gAgA_abs"])
        # Drop rows with non-numeric or zero/negative values (log scale requires positive)
        df = df.apply(pd.to_numeric, errors="coerce").dropna()
        df = df[(df["lambda_m"] > 0) & (df["gAgA_abs"] > 0)]
        datasets[name] = df
    except Exception as e:
        print(f"Skipping {name}: {e}")

print(f"Loaded {len(datasets)} datasets")
print("V2 datasets found:", [k for k in datasets if k.startswith("2")])

# ── Plot ──
fig, ax = plt.subplots(figsize=(10, 7))

colors = plt.cm.tab10.colors
color_idx = 0

for name, df in datasets.items():
    if name.startswith("2"):   # V2 potential only
        # Clean label: strip leading "2", replace underscores
        label = name[1:].replace("_m_abs_", " ").replace("_", " ")
        ax.loglog(df["lambda_m"], df["gAgA_abs"],
                  label=label,
                  color=colors[color_idx % len(colors)],
                  linewidth=1.5)
        color_idx += 1

ax.set_xlabel("Interaction range λ (m)", fontsize=12)
ax.set_ylabel("|g$_A$g$_A$|", fontsize=12)
ax.set_title("V2 g$_A$g$_A$ constraints: lepton–lepton", fontsize=13)
ax.legend(fontsize=8, loc="upper right")
ax.grid(True, which="both", ls="--", alpha=0.4)
plt.tight_layout()
plt.savefig("V2_gAgA_lepton_lepton.pdf")
plt.show()
```

```python
# Classify each dataset by sector
def get_sector(name):
    if "eebar" in name or "emu" in name:
        return "antimatter"
    elif "ee" in name:
        return "matter"
    else:
        return "other"

sector_colors = {"matter": "blue", "antimatter": "red", "other": "green"}
```

```python
# For each potential term and λ value where both sectors have data,
# interpolate and compute:
# A_alpha = (g_matter - g_antimatter) / (g_matter + g_antimatter)
# Then compute chi-squared for consistency test

from scipy.interpolate import interp1d
import numpy as np

def compute_asymmetry(df_matter, df_antimatter):
    # Find overlapping λ range
    lam_min = max(df_matter["lambda_m"].min(), df_antimatter["lambda_m"].min())
    lam_max = min(df_matter["lambda_m"].max(), df_antimatter["lambda_m"].max())
    
    lam_grid = np.logspace(np.log10(lam_min), np.log10(lam_max), 100)
    
    f_m = interp1d(df_matter["lambda_m"], df_matter["gAgA_abs"], 
                   kind="linear", bounds_error=False)
    f_a = interp1d(df_antimatter["lambda_m"], df_antimatter["gAgA_abs"], 
                   kind="linear", bounds_error=False)
    
    g_m = f_m(lam_grid)
    g_a = f_a(lam_grid)
    
    A = (g_m - g_a) / (g_m + g_a)
    return lam_grid, A
```
